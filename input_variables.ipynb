```python
# Parameters
model_path = None
num_samples = None
test_count = None
num_classes = None
X_test = None
y_test = None
```

```python
import papermill as pm

# Function to get input variables from the user
def get_user_input():
    model_path = input("Enter the model path: ")
    num_samples = int(input("Enter the number of samples: "))
    test_count = int(input("Enter the test count: "))
    num_classes = int(input("Enter the number of classes: "))
    X_test = input("Enter the path for X_test data: ")
    y_test = input("Enter the path for y_test data: ")
    return model_path, num_samples, test_count, num_classes, X_test, y_test

if __name__ == "__main__":
    model_path, num_samples, test_count, num_classes, X_test, y_test = get_user_input()

    # Define the path to the input and output notebooks
    input_notebook = 'Uncertainty_quantification.ipynb'
    output_notebook = 'Uncertainty_quantification_output.ipynb'

    # Execute the notebook with the provided parameters
    pm.execute_notebook(
        input_notebook,
        output_notebook,
        parameters={
            'model_path': model_path,
            'num_samples': num_samples,
            'test_count': test_count,
            'num_classes': num_classes,
            'X_test': X_test,
            'y_test': y_test
        }
    )
```
